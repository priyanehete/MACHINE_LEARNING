```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings("ignore")
```

```python
dt_train = pd.read_csv('train.csv')
dt_test = pd.read_csv('test.csv')
```

```python
dt_train.head()
```

```python
dt_test.head()
```

```python
print(dt_train.shape)
print(dt_test.shape)
```

```python
dt_train['Data']= 'Train'
dt_test['Data']='Test'
```

```python
dt_train.head()
```

```python
dt_test.head()
```

```python
dt_train.isnull().sum()
```

```python
dt_test.isnull().sum()
```

```python
# Train and test data are combined for preprocessing
dt = pd.concat([dt_train, dt_test], axis =0)
```

```python
dt.head()
```

```python
dt.tail()
```

```python
dt.shape
```

```python
dt.isnull().sum()/len(dt)*100
```

```python
# Dropping non-significant variable
dt.drop(['Cabin','PassengerId', 'Name', 'Ticket', 'Fare'], axis = 1,inplace =True)
```

```python
dt.head()
```

```python
dt.info()
```

```python
#treating null values
dt.isnull().sum()
```

```python
dt.describe()
```

```python
dt['Survived']=dt['Survived'].fillna(dt['Survived'].mean())
```

```python
dt['Age'].fillna(dt['Age'].median(), inplace =True)
```

```python
dt['Embarked'].value_counts()
```

```python
dt['Embarked'].fillna('S', inplace = True)
```

```python
dt.isnull().sum()
```

```python
#Encoding
#Sex, Embarked 

dt = pd.get_dummies(dt, columns = ['Sex', 'Embarked'],drop_first =True)
```

```python
dt.tail()
```

```python
dt.info()
```

```python
# Train and Test are seperated again after preprocessing

train = dt[dt['Data']=='Train']
test = dt[dt['Data']=='Test']
print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
test = test.drop(['Survived','Data'], axis = 1)
```

```python
test.head()
```

```python
# dpenedent and independent variable from train data

x = train.drop(['Survived', 'Data'], axis = 1)
y = train[['Survived']]
```

```python
x.head()
```

```python
y.head()
```

```python
train['Survived'].value_counts()
```

<br>

<br>

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#split the data into train & test from training data to build the model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size  = 0.25, random_state =102)
```

```python
#building SVM Model
svm_linear = SVC(kernel = 'linear')
svm_linear.fit(x_train, y_train)
```

```python
#prediction
y_svm_linear_train_predict = svm_linear.predict(x_train)
y_svm_linear_test_predict = svm_linear.predict(x_test)
```

```python
train_accuracy_score = accuracy_score(y_train, y_svm_linear_train_predict)
test_accuracy_score = accuracy_score(y_test, y_svm_linear_test_predict )

print(f'Train Accuracy: {train_accuracy_score}\nTest Accuracy: {test_accuracy_score}')
```

```python
#predict survived with test data with model svm LInear
y_svm_linear_test_data_predict = svm_linear.predict(test)
```

```python
test['Predicted Survived']= y_svm_linear_test_data_predict 
```

```python
test.head()
```

```python
test_out = pd.merge(test, test[['Predicted Survived']], how='left', left_index = True, right_index =True )
```

```python
test_out
```

```python
#Cross Validation
cv = cross_val_score(svm_linear, x_train, y_train, cv = 10)
print(cv.mean())
print(cv.max())
```
