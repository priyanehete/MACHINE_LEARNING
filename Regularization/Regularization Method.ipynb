```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

from sklearn.metrics import r2_score
```

```python
mpg_df = pd.read_csv('mpg.csv')
mpg_df.head()
```

```python
# drop name
mpg_df = mpg_df.drop('name', axis=1)
mpg_df.head()
```

```python
mpg_df['origin'].value_counts()
```

```python
mpg_df['origin'] = mpg_df['origin'].replace({1:'america',2:'europe',3:'asia'})
mpg_df.head()
```

```python
mpg_df['origin'].value_counts()
```

```python
# One hot encoder method
mpg_df = pd.get_dummies(mpg_df, columns=['origin'])
```

```python
mpg_df.head()
```

```python
# dummy variable : n-1
# dropping origin_america
mpg_df = mpg_df.drop('origin_america', axis=1)
```

```python
mpg_df.head()
```

```python
mpg_df.isnull().sum()
```

```python
# Finding unique values in the columns
for i in mpg_df.columns:
    print("******************************************************", i,
         "**************************************************************")
    print()
    print(set(mpg_df[i].tolist()))
    print()
```

```python
mpg_df = mpg_df.replace('?', np.nan)
mpg_df = mpg_df.apply(lambda x : x.fillna(x.median()), axis=0)
```

```python
# Seperate independent variable and dependent variable
x = mpg_df.drop('mpg', axis=1)
y = mpg_df[['mpg']]
```

```python
x.head()
```

```python
y.head()
```

```python
# pls treate outlier as well, in case found

# feature scaling
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler()
x_scaled = x_scaled.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
x_scaled
```

```python
# Split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=1)
```

# Building Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
linear = LinearRegression()
linear.fit(x_train, y_train)

for idx, col_name in enumerate(x_train.columns):
    print("The coefficient for {} is {}".format(col_name, linear.coef_[0][idx]))
```

```python
intercept = linear.intercept_[0]
print("the intercept for our linear model is {}".format(intercept))
```

# Regularisation Method
## Part 1 - Lasso model 

```python
from sklearn.linear_model import Lasso 
# exact zero or closure to zero
# penalty = 0.1
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
print("Lasso Model :", (lasso.coef_))
```

```python
y_pred_train_lasso = lasso.predict(x_train)
y_pred_test_lasso = lasso.predict(x_test)
```

```python
print("Training Accuracy :", r2_score(y_train, y_pred_train_lasso))
print()
print("Test Accuracy :", r2_score(y_test, y_pred_test_lasso))
```

```python
lasso.intercept_
```

```python
# Part 2 : Ridge Regression (L2- Regularization)
# closure to zero but not exact zero
# penalty - 0.3
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.3)
ridge.fit(x_train, y_train)
print("Ridge Model :", (ridge.coef_))
```

```python
y_pred_train_ridge = ridge.predict(x_train)
y_pred_test_ridge = ridge.predict(x_test)
```

```python
print("Training Accuracy :", r2_score(y_train, y_pred_train_ridge))
print()
print("Test Accuracy :", r2_score(y_test, y_pred_test_ridge))
```
